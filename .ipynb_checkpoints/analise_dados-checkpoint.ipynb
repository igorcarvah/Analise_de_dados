```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def plot_histogram_with_kde(
    data,
    bins=20,
    bar_color="skyblue",
    kde_color="red",
    title="Histograma com KDE",
    xlabel="Valores",
    ylabel="Densidade"
):
    """
    Plota um histograma com curva de densidade KDE sobreposta.

    Parâmetros:
    - data: array-like ou Series com os dados.
    - bins: número de divisões do histograma.
    - bar_color: cor das barras do histograma.
    - kde_color: cor da curva KDE.
    - title: título do gráfico.
    - xlabel: rótulo do eixo X.
    - ylabel: rótulo do eixo Y.
    """

    plt.figure(figsize=(10, 6))

    # Histograma
    sns.histplot(
        data,
        bins=bins,
        stat="density",        # escala compatível com a KDE
        color=bar_color,
        edgecolor="black",     # borda nas barras
        alpha=0.7              # transparência para melhorar visualização
    )

    # Curva KDE
    sns.kdeplot(
        data,
        color=kde_color,
        linewidth=2,
        label="KDE"
    )

    # Customização visual
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Exemplo de uso
medias_por_filme = pd.Series([2.5, 3.0, 4.0, 3.8, 2.9, 4.2, 4.8, 3.3, 3.1, 3.7])
plot_histogram_with_kde(
    data=medias_por_filme,
    bins=10,
    bar_color="skyblue",
    kde_color="red",
    title="Distribuição das Médias dos Filmes",
    xlabel="Média de Avaliações",
    ylabel="Densidade"
)

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Carrega os dados
tmdb = pd.read_csv('tmdb_5000_movies.csv', low_memory=False)

# Agrupa por idioma e calcula a média, incluindo orçamentos zerados
df_media = (
    tmdb
    .groupby('original_language', as_index=False)['budget']
    .mean()
    .rename(columns={'budget': 'budget_medio'})
    .sort_values(by='budget_medio', ascending=False)
)

# Adiciona coluna com log10 do orçamento médio
# Usa .replace(0, 1) para evitar erro de log(0), só para o gráfico
df_media['budget_log'] = df_media['budget_medio'].replace(0, 1).apply(np.log10)

# Plota o gráfico
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_media,
    x='budget_log',
    y='original_language',
    palette='viridis',
    edgecolor='black'
)

plt.xlabel('Log10 do Orçamento Médio por Filme (USD)')
plt.ylabel('Língua')
plt.title('Orçamento Médio por Filme e por Língua (incluindo idiomas com orçamento zero)')
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Remove orçamentos zerados (opcional, mas importante para média)
tmdb = tmdb[tmdb['budget'] > 0]

# Agrupa por idioma e calcula a média
df_media = (
    tmdb
    .groupby('original_language', as_index=False)['budget']
    .mean()
    .rename(columns={'budget': 'budget_medio'})
    .sort_values(by='budget_medio', ascending=False)
)

# Adiciona coluna log10 da média
df_media['budget_log'] = df_media['budget_medio'].apply(np.log10)

# Plota o gráfico
plt.figure(figsize=(12, 6))
sns.barplot(
    data=df_media,
    x='budget_log',
    y='original_language',
    palette='viridis',
    edgecolor='black'
)

plt.xlabel('Log10 do Orçamento Médio por Filme')
plt.ylabel('Língua')
plt.title('Orçamento Médio por Filme e por Língua (escala log no orçamento)')
plt.grid(axis='x', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
